Load libraries and dataset

```python
import pandas as pd
import os

gold_path = "integrated_datasets/gold_integrated.csv"
gold_df = pd.read_csv(gold_path)

print("Gold dataset shape:", gold_df.shape)
gold_df.head()
```

Check missing values

```python
missing_percent = gold_df.isnull().mean() * 100
missing_percent.sort_values(ascending=False)
```

Remove irrelevant metadata columns

```python
irrelevant_cols = [
    'Source_Campus_File',
    'Source_Campus_File_x',
    'Source_Campus_File_y',
    'Dataset_Type',
    'Dataset_Type_x',
    'Dataset_Type_y'
]

gold_df.drop(
    columns=[c for c in irrelevant_cols if c in gold_df.columns],
    inplace=True
)

print("Irrelevant metadata removed.")
print("Shape:", gold_df.shape)
```

Remove non-analytical identifier

```python
gold_df.drop(columns=['Full_Name'], inplace=True, errors='ignore')
print("Name column removed.")
# If Campus_Name_x and Campus_Name_y are identical, keep one
gold_df['Campus_Name'] = gold_df['Campus_Name_x'].combine_first(gold_df['Campus_Name_y'])

# Drop the original duplicated columns
gold_df.drop(columns=['Campus_Name_x', 'Campus_Name_y'], inplace=True)

```

Final dataset check

```python
gold_df.info()
gold_df.head()
```

Save reduced dataset

```python

os.makedirs("reduced_datasets", exist_ok=True)

reduced_path = "reduced_datasets/gold_reduced.csv"
gold_df.to_csv(reduced_path, index=False)

print("Reduced dataset saved:", reduced_path)
```
